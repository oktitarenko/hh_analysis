3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
#Функция для преобразования опыта в месяцы
def convert_experience_to_months(text):
    if pd.isna(text) or text == "Не указано":
        return np.nan
    
    # Убираем "Опыт работы: " и обрабатываем остаток
    text = text.replace("Опыт работы: ", "").split(",")[0]  # Оставляем только первую часть строки

    years = 0
    months = 0
    
    # Ищем информацию о годах
    if 'год' in text or 'года' in text or 'лет' in text:
        words = text.split()
        for i in range(len(words)):
            if words[i] in ['год', 'года', 'лет']:  # Если нашли слово, связанное с годами
                if i > 0 and words[i-1].isdigit():
                    years = int(words[i-1])  # Извлекаем число перед словом "год" или "лет"
    
    # Ищем информацию о месяцах
    if 'месяц' in text or 'месяца' in text or 'месяцев' in text:
        words = text.split()
        for i in range(len(words)):
            if words[i] in ['месяц', 'месяца', 'месяцев']:  # Если нашли слово, связанное с месяцами
                if i > 0 and words[i-1].isdigit():
                    months = int(words[i-1])  # Извлекаем число перед словом "месяц" или "месяцев"
    
    # Переводим все в месяцы (1 год = 12 месяцев)
    total_months = years * 12 + months
    return total_months


# Применяем функцию к столбцу и создаем новый столбец "Опыт работы (месяц)"
df["Опыт работы (месяц)"] = df["Опыт работы"].apply(convert_experience_to_months)
df = df.drop("Опыт работы", axis=1)
```

```python
million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 
                  'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград']


# Функции для извлечения признаков
def extract_city(text):
    """Функция для извлечения города и классификации его по категориям."""
    city = text.split(",")[0].strip()  # Извлекаем город, который перед первым запятой
    if city == "Москва":
        return "Москва"
    elif city == "Санкт-Петербург":
        return "Санкт-Петербург"
    elif city in million_cities:
        return "город-миллионник"
    else:
        return "другие"

def extract_relocation(text):
    """Функция для извлечения готовности к переезду."""
    if "готов к переезду" in text or "хочу переехать" in text:
        return True
    elif "не готов" in text:
        return False
    return False  # Если информация о переезде отсутствует, считаем, что не готов

def extract_business_trip(text):
    """Функция для извлечения готовности к командировкам."""
    if "готов к командировкам" in text or "готова к командировкам" in text:
        return True
    elif "не готов" in text:
        return False
    return False  # Если информация о командировках отсутствует, считаем, что не готов

# Применяем функции для извлечения признаков
df['Город'] = df['Город, переезд, командировки'].apply(extract_city)
df['Готовность к переезду'] = df['Город, переезд, командировки'].apply(extract_relocation)
df['Готовность к командировкам'] = df['Город, переезд, командировки'].apply(extract_business_trip)

df = df.drop('Город, переезд, командировки', axis=1)
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# Функция для выполнения One-Hot Encoding
categories_occupy = ['полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка']
categories_schedule = ['полный день', 'сменный график', 'гибкий график', 'удаленная работа', 'вахтовый метод']

def one_hot_encode(text, categories):
    # Инициализируем пустой словарь с нулями
    encoded = {category: 0 for category in categories}
    
    # Разбиваем строку на категории и обновляем словарь
    for category in text.split(','):
        category = category.strip()  # Убираем лишние пробелы
        if category in encoded:
            encoded[category] = 1  # Если категория присутствует, ставим 1
    return pd.Series(encoded)

df_occupy = df['Занятость'].apply(one_hot_encode, categories=categories_occupy)
df_schedule = df['График'].apply(one_hot_encode, categories=categories_schedule)
df = df.drop(['Занятость', 'График'], axis=1)
df = pd.concat([df, df_occupy, df_schedule])
```

```python
currencies = pd.read_csv('ExchangeRates.csv')

# Преобразуем столбец "Обновление резюме" в формат datetime
df['Обновление резюме'] = pd.to_datetime(df['Обновление резюме']).dt.date
```

```python
# Словарь для перевода валют в ISO код
to_iso = {
    'грн': 'UAH',
    'USD': 'USD',
    'EUR': 'EUR',
    'белруб': 'BYN',
    'KGS': 'KGS',
    'сум': 'UZS',
    'AZN': 'AZN',
    'KZT': 'KZT'
}

# Функция для извлечения суммы и валюты
def extract_salary_and_currency(salary_str):
    # Убедимся, что значение в строковом формате
    salary_str = str(salary_str)
    
    # Разделяем строку на сумму и валюту
    parts = salary_str.split()
    
    # Если строка разделилась на 2 части (сумма и валюта), то извлекаем их
    if len(parts) == 2:
        try:
            value = float(parts[0])  # сумма
            currency = parts[1].lower()  # валюта в нижнем регистре
            # Переводим валюту в ISO код
            iso_currency = to_iso.get(currency, currency)  # если валюта не найдена, оставляем как есть
            return pd.Series([value, iso_currency])
        except ValueError:
            return pd.Series([None, None])  # Если не удается преобразовать сумму в float, возвращаем None
    else:
        return pd.Series([None, None])

# Применяем функцию для выделения суммы и валюты
df[['value', 'currency']] = df['ЗП'].apply(extract_salary_and_currency)
```

Напомним, сделать это можно с помощью метода axvline. Например, для построение линии среднего будет иметь вид:

`histplot.axvline(log_age.mean(), color='k', lw=2)`

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
# 1. Преобразуем возраст в логарифмическую шкалу
log_age = np.log(df_cleaned['Возраст'])

# 2. Построение гистограммы
plt.figure(figsize=(10, 6))
sns.histplot(log_age, bins=30, kde=True)

# 3. Добавление линии среднего
plt.axvline(log_age.mean(), color='k', lw=2, label='Среднее')

# 4. Добавление границ интервала трех сигм (4 сигмы для правой стороны)
std_dev = log_age.std()
mean = log_age.mean()

# Границы интервала - 4 сигмы справа
upper_bound = mean + 4 * std_dev

# Линии для интервала
plt.axvline(upper_bound, color='r', linestyle='--', label='4 сигмы (верхняя граница)')

# 5. Заголовок и легенда
plt.title("Распределение возраста в логарифмическом масштабе с границами интервала 4 сигм", fontsize=14)
plt.legend()
plt.show()

# 6. Находим выбросы (по Z-отклонению в логарифмическом масштабе)
z_scores = (log_age - mean) / std_dev
outliers = df_cleaned[z_scores > 4]  # выбросы правой стороны, с послаблением на 4 сигмы

# 7. Выводим таблицу с выбросами
print(outliers[['Возраст', 'ЗП_число', 'Опыт работы (месяц)']])
```
